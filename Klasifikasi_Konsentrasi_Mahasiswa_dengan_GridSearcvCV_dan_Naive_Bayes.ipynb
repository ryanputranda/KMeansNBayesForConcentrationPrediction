```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
#file_path = 'path/to/your/datasetclustering.xlsx'
#df = pd.read_excel(r'datasetclustering.xls')

df = pd.read_excel(r'/content/dfuntukklasifikasi.xlsx')
df

```

```python
# prompt: hapus kolom technopreneurship_encode dan cluster_encode

# Drop the specified columns
df = df.drop(['technopreneurship_encode', 'cluster_encode'], axis=1)

# Display the updated DataFrame (optional)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding untuk kolom 'diagnosa'
le_technopreneurship = LabelEncoder()
le_cluster = LabelEncoder()

df['technopreneurship_encode'] = le_technopreneurship.fit_transform(df['technopreneurship'])
df['cluster_encode'] = le_cluster.fit_transform(df['cluster'])
```

```python
df.head()
```

```python
# Pisahkan fitur (X) dan target (y)
x = df[['pemrograman?', 'matematis statistik?', 'tata kelola SI/TI?', 'multimedia?', 'seni dalam multimedia?', 'analitis sistematis?', 'memimpin?', 'berpikir kreatif?', 'technopreneurship_encode']]
y = df[['cluster_encode']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
# prompt: adopsi naive bayes untuk prediksinya

from sklearn.naive_bayes import GaussianNB

# Inisialisasi model Naive Bayes
model = GaussianNB()

# Latih model dengan data training
model.fit(x_train, y_train.values.ravel())

# Prediksi dengan data testing
y_pred = model.predict(x_test)

# Evaluasi model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# prompt: lakukan prediksi pada data testing 10 baris terakhir secara acak dan bandingkan antara atribut cluster aslinya dengan prediksi modelnya

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.naive_bayes import GaussianNB
import random

# Select 10 random rows from the test set
random_indices = random.sample(range(len(x_test)), 10)
x_test_random = x_test.iloc[random_indices]
y_test_random = y_test.iloc[random_indices]

# Make predictions on the random subset
y_pred_random = model.predict(x_test_random)

# Compare predictions with actual values
comparison_df = pd.DataFrame({'Actual Cluster': y_test_random['cluster_encode'], 'Predicted Cluster': y_pred_random})
comparison_df
```

# **ADOPSI GRIDSEARCHCV UNTUK HYPERPARAMETER TUNNING OPTIMIZATIONNYA**

```python
# prompt: gunakan gridsearchcv untuk naive bayes

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score
from sklearn.naive_bayes import GaussianNB
import random

# Initialize Gaussian Naive Bayes model
model = GaussianNB()

# Define the parameter grid for GridSearchCV
param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5] # Example values, adjust as needed
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy') # You can change scoring metric if needed
grid_search.fit(x_train, y_train.values.ravel())


# Get the best model and its hyperparameters
best_model = grid_search.best_estimator_
print("Best Hyperparameters:", grid_search.best_params_)

# Make predictions using the best model
y_pred = best_model.predict(x_test)

# Evaluate the best model
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Random predictions example (as in the original code)
random_indices = random.sample(range(len(x_test)), 10)
x_test_random = x_test.iloc[random_indices]
y_test_random = y_test.iloc[random_indices]
y_pred_random = best_model.predict(x_test_random)

comparison_df = pd.DataFrame({'Actual Cluster': y_test_random['cluster_encode'], 'Predicted Cluster': y_pred_random})
comparison_df
```
